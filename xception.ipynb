```python
#Import Numpy, Keras Image and Xception libraries
import numpy as np
from keras.preprocessing import image
from tensorflow.keras.applications import xception
from tensorflow.keras.applications.xception import Xception
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
model = Xception(weights='imagenet')
model
```

```python
img = image.load_img('sealion.jpg', target_size=(299,299))
img
```

```python
#convert image to array
input_img = image.img_to_array(img)
input_img = np.expand_dims(input_img, axis=0)
input_img = xception.preprocess_input(input_img)
input_img
```

```python
predict_img = model.predict(input_img)
predict_img
```

```python
top_five_predict = xception.decode_predictions(predict_img, top=5)
top_five_predict
```
